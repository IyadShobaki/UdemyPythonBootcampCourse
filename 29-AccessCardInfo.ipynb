```python
class Card:
 
    """ASK THE ACCESS CARD INFORMATIONS TO THE CLIENT AND CHECK IF THE ACCESS CARD IS VALID"""
 
    def __init__(self):
 
        """ Card constructor 'name' 'number' 'date' """
 
        self.name = input("Please enter your name: ")
        self.number = input("Please enter your access card number: ")
        self.date = input("Please enter the date when you card will expire: ")
        print()
 
 
    def __str__(self):
 
        """Display the access card number information input"""
 
        return f"Your name: {self.name}\nYour access card: n°{self.number}\nYour access card will expire: {self.date}\n"
 
    def confirm_info(self):
 
        """Ask to the client to confirm the information, if not ask to input again"""
 
        info_checked = False
 
        while info_checked is False:
 
            confirm = input(f"Do you confirm your informations?: ")
            print()
 
            if confirm[0].lower() == "y":
                info_checked = True
            else:
                self.name = input("Please enter your name: ")
                self.number = input("Please enter your access card number: ")
                self.date = input("Please enter the date when you card will expire: ")
                info_checked = False
                print()
 
    def confirm_nb_card(self):
 
        """Control the access card number for validation"""
 
        number = self.number[::-1]
        total = 0
        for i in number[1::2]:
            x = int(i) * 2
            if len(str(x)) == 2:
                for a in str(x):
                    total += int(a)
            else:
                total += int(x)
 
        for i in number[::2]:
            total += int(i)
 
        return total
 
    def test_card(self):
 
        """ Validation of the type of card or not """
 
        number = self.number[0::]
        if (int(number[:2]) >= 51 and int(number[:2]) <= 55 and len(number) == 16):
            print(f"{self.name} your access card n°{self.number} is a MasterCard.")
            print()
            r = input("Would you like to check a second access card?: ")
            print()
            if r[0].lower() == "n":
                return False
 
        elif (int(number[:2]) == 34 and int(number[:2]) == 37 and len(number) == 15):
            print(f"{self.name} your access card n°{self.number} is a American Express Card.")
            print()
            r = input("Would you like to check a second access card?: ")
            print()
            if r[0].lower() == "n":
                return False
 
        elif (int(number[0]) == 4 and len(number) == 13) or (int(number[0]) == 4 and len(number) == 16):
            print(f"{self.name} your access card n°{self.number} is a Visa Card.")
            print()
            r = input("Would you like to check a second access card?: ")
            print()
            if r[0].lower() == "n":
                return False
        else:
            print("Sorry, your information don't match with our record...")
            print()
            r = input("Would you like to check a second access card?: ")
            print()
            if r[0].lower() == "n":
                return False
 
```

```python
process_asked = True
 
while process_asked is True:
 
    mycard = Card()
    print(mycard)
    mycard.confirm_info()
    mycard.confirm_nb_card()
 
    if mycard.test_card() == False:
        print("Thank you, have a good day!")
        break
```
